```python
import numpy as np
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
age = [17.5, 22, 29.5, 44.5, 64.5, 80]
deaths = [38, 36, 24, 20, 18, 28]
```

```python
plt.plot( age, deaths,'r-')
```

```python
plt.scatter(age, deaths)
plt.ylabel("Number of Driver Deaths per 100,000")
plt.xlabel("Age")
plt.show()
```

```python
N = 6
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2 
```

```python
plt.scatter(age, deaths, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
X_b = np.c_[np.ones((6, 1)), age]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(deaths)
```

```python
theta_best
```

```python
X_new = np.array([[15], [90]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(age, deaths, "g.")
plt.axis([15, 90, 0, 40])
plt.show()
```
